```python
import cv2
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# load image and find contours
img = cv2.imread("example0.png", cv2.IMREAD_GRAYSCALE)
img = filters.median_filter(img.astype('uint8'), size=15)
plt.imshow(img)
contours, _ = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)


(x,y),radius = cv2.minEnclosingCircle(contours[0])
center = (int(x),int(y))

# get contour points in polar coordinates
rhos = []
phis = []
for i in range(len(contours[0])):
    x, y = contours[0][i][0]
    rho, phi = cart2pol(x-center[0], y-center[1])
    rhos.append(rho)
    phis.append(phi)

rhos = smooth(rhos, 5)
print("peaks", find_peaks(rhos, height=0))
n_knobs = len(find_peaks(rhos, height=0)[0]) - 4

print(f"KNOBS: {n_knobs}")
print(rhos.mean())

plt.show()
plt.plot(rhos)
```

```python
def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)

def smooth(y, box_pts):
    box = np.ones(box_pts)/box_pts
    y_smooth = np.convolve(y, box, mode='valid')
    return y_smooth
```

```python
import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import filters

def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return (rho, phi)

img = cv2.imread('example.png', cv2.IMREAD_GRAYSCALE)
img = filters.median_filter(img.astype('uint8'), size=18)

plt.imshow(img)
plt.show()

contours, _ = cv2.findContours(img, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

M = cv2.moments(contours[0])
cx = int(M["m10"] / M["m00"])
cy = int(M["m01"] / M["m00"])

polar = [cart2pol(c[0][0] - cx, c[0][1] - cy) for c in contours[0][:]]
max_i = polar.index(max(polar, key = lambda x: x[1]))
polar = polar[max_i:] + polar[:max_i]
ds, phis = zip(*polar)

plt.gcf().set_size_inches(6, 3)     
plt.plot(phis, ds)
plt.show()
```
